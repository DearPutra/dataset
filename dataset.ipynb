```python
import pandas as pd
```

```python
data1 = pd.read_csv("DoS UDP Flood.csv")
```

```python
data1
```

```python
data2 = pd.read_csv("MITM ARP Spoofing.csv")
```

```python
data2
```

```python
data3 = pd.read_csv("MQTT DoS Publish Flood.csv")
```

```python
data3
```

```python
gab = pd.concat([data1,data2,data3],ignore_index=True)
```

```python
gab.columns.values
```

```python
x = gab.iloc[:,7: 76]
```

```python
x
```

```python
y = gab.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
desc = DecisionTreeClassifier(criterion = 'entropy', splitter = 'random')
desc.fit(x_train,y_train)
```

```python
y_pred = desc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
acy = accuracy_score(y_test,y_pred)
```

```python
acy
```

```python
import matplotlib.pyplot as pyp
import numpy as np
```

```python
fgr = pyp.figure(figsize = (10, 7))
tree.plot_tree(desc, feature_names = x.columns.values, class_names = np.array(['DoS UDP Flood','MITM ARP Spoofing','MQTT DoS Publish Flood']), filled = True)
pyp.show()
```

```python
import seaborn as sea
from sklearn import metrics
label = np.array(['DoS UDP Flood','MITM ARP Spoofing','MQTT DoS Publish Flood'])
```

```python
import matplotlib.pyplot as pyp
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
pyp.figure(figsize=(10, 10))
sea.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
pyp.xlabel('Prediksi')
pyp.ylabel('Fakta')
pyp.show()
```
